# Part 4 : Visualise training data on a 2-dimensional plot
<font color = blue>Here I have shown 36 plots which corresponds to all possible combinations of 9 attributes in training data.</blue>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations 

data = pd.read_csv("train.csv")

columns = ["promotion_last_5years", "Work_accident", "sales", "salary","satisfaction_level","last_evaluation","number_project","average_montly_hours","time_spend_company"]
all_possible = combinations(columns, 2)
for i in list(all_possible):
    fig, axes = plt.subplots(figsize=(12,6))
    positive = data[data['left'] == 0]
    x = positive[i[0]].values 
    y = positive[i[1]].values 
    axes.scatter(x, y, label= "Employees Not Left", color= "blue",marker= "*", s=30) 

    negative = data[data['left'] == 1]
    x = negative[i[0]].values 
    y = negative[i[1]].values 
    axes.scatter(x, y, label= "Employees Left", color= "red",marker= "*", s=30)

    plt.xlabel(i[0]) 
    plt.ylabel(i[1]) 
    plt.title(i[0]+ " vs "+ i[1]) 
    plt.legend() 
    name=i[0]+i[1]+".png"
    plt.savefig(name,bbox_inches='tight')
    plt.show() 
```
